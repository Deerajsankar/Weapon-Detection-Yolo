```python
from ultralytics import YOLO

# Load YOLO model
model = YOLO(r"/content/yolo11n.pt")  # Ensure the model file exists in your working directory

# Train the model
results = model.train(
    data="/content/Weapon-Detection-1/data.yaml",  # Path to your data.yaml
    epochs=25,  # Number of epochs
    batch=16,   # Batch size
    imgsz=911   # Image size
)
```

```python
from ultralytics import YOLO

# Load the best model weights
model = YOLO("/content/runs/detect/train2/weights/best.pt")

# Evaluate the model on the validation dataset
metrics = model.val()  # This will use the validation dataset specified in data.yaml
print(metrics.box.map)  # mAP50-95
print(metrics.box.map50)  # mAP50
print(metrics.box.map75)  # mAP75
```
